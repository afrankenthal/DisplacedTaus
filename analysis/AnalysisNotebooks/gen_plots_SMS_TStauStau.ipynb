```python
import ROOT
```

```python
d = ROOT.RDataFrame("ntuples/genT", "../merged_SMS_TStauStau_LH_mStau-225to250.root")
```

```python
# S = Stau
d_S = d.Define("S_mask", "abs(gen_ID) == 1000015")
h_pt_S = d_S.Define("pt_S", "gen_pt[S_mask]").Histo1D("pt_S")
h_deta_S = d_S.Define("deta_S", "gen_eta[S_mask][0] - gen_eta[S_mask][1]").Histo1D("deta_S")
h_red_dphi_S = d_S.Define("dphi_S", "abs(gen_phi[S_mask][0] - gen_phi[S_mask][1])").\
    Define("red_dphi_S", "abs(dphi_S - 3.141592 * (1 + (dphi_S-3.141592)/abs(dphi_S-3.141592)))").\
    Histo1D("red_dphi_S")
h_mass_S = d_S.Define("mass_Stau", "gen_mass[S_mask]").Histo1D("mass_Stau")
```

```python
c = ROOT.TCanvas()
h_mass_S.Draw()
h_mass_S.GetXaxis().SetTitle("mass(Stau) [GeV]")
c.Draw()
```

```python
# LSP
d_LSP = d.Define("LSP_mask", "abs(gen_ID) == 1000022")
h_pt_LSP = d_LSP.Define("pt_LSP", "gen_pt[LSP_mask]").Histo1D("pt_LSP")
h_deta_LSP = d_LSP.Define("deta_LSP", "gen_eta[LSP_mask][0] - gen_eta[LSP_mask][1]").Histo1D("deta_LSP")
h_red_dphi_LSP = d_LSP.Define("dphi_LSP", "abs(gen_phi[LSP_mask][0] - gen_phi[LSP_mask][1])").\
    Define("red_dphi_LSP", "abs(dphi_LSP - 3.141592 * (1 + (dphi_LSP-3.141592)/abs(dphi_LSP-3.141592)))").\
    Histo1D("red_dphi_LSP")
h_mass_LSP = d_LSP.Define("mass_LSP", "gen_mass[LSP_mask]").Histo1D("mass_LSP")
```

```python
c = ROOT.TCanvas()
h_mass_LSP.Draw()
h_mass_LSP.GetXaxis().SetTitle("mass(LSP) [GeV]")
c.Draw()
```

```python
d_tau = d.Define("tau_mask", "abs(gen_ID) == 15").Define("S_mask", "abs(gen_ID) == 1000015")\
        .Define("LSP_mask", "abs(gen_ID) == 1000022")
d_tau = d_tau.Define("deta_tau01", "abs(gen_eta[tau_mask][0] - gen_eta[tau_mask][1])")
d_tau = d_tau.Define("dphi_tau01", "abs(gen_phi[tau_mask][0] - gen_phi[tau_mask][1])")
d_tau = d_tau.Define("red_dphi_tau01", "abs(dphi_tau01 - 3.141592 * (1 + (dphi_tau01-3.141592)/abs(dphi_tau01-3.141592)))")
d_tau = d_tau.Define("dR_tau01", "sqrt(pow(deta_tau01,2) + pow(red_dphi_tau01,2))")
```

```python
h_dphi_tau01 = d_tau.Histo1D("red_dphi_tau01")
```

```python
c = ROOT.TCanvas()
h_dphi_tau01.Draw()
h_dphi_tau01.GetXaxis().SetTitle("|d#phi(#tau_{i}, #tau_{j})|")
c.Draw()
c.BuildLegend(0.6, 0.4, 0.8, 0.7)
```

```python
d_tau = d_tau.Define("pt_tau_0", "gen_pt[tau_mask][0]")
d_tau = d_tau.Define("pt_tau_1", "gen_pt[tau_mask][1]")
h_pt_tau_0 = d_tau.Filter('gen_mass[S_mask][0]>240').Filter('gen_mass[LSP_mask][0]<5').Histo1D(("pt_tau_0", "pt_tau_0", 100, 0., 800.), "pt_tau_0")
h_pt_tau_1 = d_tau.Filter('gen_mass[S_mask][0]>240').Filter('gen_mass[LSP_mask][0]<5').Histo1D(("pt_tau_1", "pt_tau_1", 100, 0., 800.), "pt_tau_1")
```

```python
c = ROOT.TCanvas()
h_pt_tau_0.Draw()
h_pt_tau_0.GetXaxis().SetTitle("p_{T} (#tau_{i}) [GeV]")
h_pt_tau_1.Draw("SAME")
h_pt_tau_1.SetLineColor(ROOT.kRed)
c.SetLogy()
c.Draw()
c.BuildLegend(0.6, 0.4, 0.8, 0.7)

c2 = ROOT.TCanvas()
h_pt_tau_0.Draw("")
h_pt_tau_0.GetXaxis().SetTitle("p_{T} (#tau_{i}) [GeV]")
h_pt_tau_1.Draw("SAME")
h_pt_tau_1.SetLineColor(ROOT.kRed)
c2.Draw()
c2.BuildLegend(0.6, 0.4, 0.8, 0.7)
```

```python
d_lsp = d.Define("tau_mask", "abs(gen_ID) == 15").Define("S_mask", "abs(gen_ID) == 1000015")\
        .Define("lsp_mask", "abs(gen_ID) == 1000022")
d_lsp = d_lsp.Define("deta_lsp", "abs(gen_eta[lsp_mask][0] - gen_eta[lsp_mask][1])")
d_lsp = d_lsp.Define("dphi_lsp", "abs(gen_phi[lsp_mask][0] - gen_phi[lsp_mask][1])")
d_lsp = d_lsp.Define("red_dphi_lsp", "abs(dphi_lsp - 3.141592 * (1 + (dphi_lsp-3.141592)/abs(dphi_lsp-3.141592)))")
d_lsp = d_lsp.Define("dR_lsp", "sqrt(pow(deta_lsp,2) + pow(red_dphi_lsp,2))")
d_lsp = d_lsp.Define("pt_diff_lsp", "gen_pt[lsp_mask][0] - gen_pt[lsp_mask][1]")
```

```python
h_red_dphi_lsp_small = d_lsp.Filter('gen_mass[S_mask][0]<240').Filter('gen_mass[lsp_mask][0]>15').Histo1D("red_dphi_lsp")
h_red_dphi_lsp_large = d_lsp.Filter('gen_mass[S_mask][0]>240').Filter('gen_mass[lsp_mask][0]>15').Histo1D("red_dphi_lsp")
```

```python
c = ROOT.TCanvas()
h_red_dphi_lsp_small.Draw()
h_red_dphi_lsp_small.SetTitle("dphi_lsp_Stau_250")
h_red_dphi_lsp_small.GetXaxis().SetTitle("d#phi(LSP_{1}, LSP_{2})")
h_red_dphi_lsp_large.Draw("SAME")
h_red_dphi_lsp_large.SetLineColor(ROOT.kRed)
h_red_dphi_lsp_large.SetTitle("dphi_lsp_Stau_225")
c.SetLogy()
c.Draw()
c.BuildLegend(0.6, 0.4, 0.8, 0.7)
```

```python
h_pt_diff_lsp_small = d_lsp.Filter('gen_mass[S_mask][0]<240').Filter('gen_mass[lsp_mask][0]>15').Histo1D("pt_diff_lsp")
h_pt_diff_lsp_large = d_lsp.Filter('gen_mass[S_mask][0]>240').Filter('gen_mass[lsp_mask][0]>15').Histo1D("pt_diff_lsp")
```
